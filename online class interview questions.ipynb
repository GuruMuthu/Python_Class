```python
st = input("string")
 
freq = {}
for i in st:
    if i in freq:
        print("true",freq)
        freq[i] += 1
    else:
        print("false",freq)
        freq[i] = 1
result = max(freq , key = freq.get)
print(result)
```

```python
num=[43,2,9,1,3,4,5,33,21,54,6,7,8]
num.sort()
for i in range(len(num)):
    f=num.pop()
    if(i==0):
        print(i+1,"'st biggest number",f)
    elif(i==1):
        print(i+1,"'nd biggest number",f)
    elif(i==2):
        print(i+1,"'rd biggest number",f)
    else:
        print(i+1,"'th biggest number",f)
```
