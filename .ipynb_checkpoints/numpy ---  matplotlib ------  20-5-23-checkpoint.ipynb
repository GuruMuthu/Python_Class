```python
import matplotlib.pyplot as plt
import numpy as np

x1point = np.array([2,5,5,2,2,5])
y1point = np.array([20,20,30,30,50,50])

plt.plot(x1point,y1point,linestyle="dashed",linewidth=5)

plt.xlabel("year ----- >")
plt.ylabel("salary  ----- >")
plt.title("employ table")
plt.grid(axis="x")
plt.show()
```

```python
# plot : 1
x1point = np.array([2,5,5,2,2,5])
y1point = np.array([20,20,30,30,50,50])

plt.subplot(2,1,2)
plt.plot(x1point,y1point,linewidth=2)
plt.xlabel("nothing")
plt.ylabel("nothing")

# plot : 2
x1point = np.array([1,2,3,4,5])
y1point = np.array([10,20,30,40,50])

plt.subplot(2,1,1)
plt.plot(x1point,y1point,linewidth=2)
plt.xlabel("nothing")

plt.show()
```

```python
x1point = np.array([1,2,3,4,5])
y1point = np.array([10,20,30,40,50])

plt.pie(x1point)
plt.xlabel("year ----- >")
plt.ylabel("salary  ----- >")
plt.title("employ table")

plt.show()
```

```python
x = np.array([25,25,25,25])
mylabels = ["apple","mango","orange","banana"]

plt.pie(x, labels= mylabels)

plt.show()
```

```python
x = np.array([25,35,25,10])
mylabels = ["apple","mango","orange","banana"]
myexplode = [0.2,0,0,0]

plt.pie(x, labels= mylabels, explode =myexplode )

plt.show()
```
